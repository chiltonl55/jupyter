## (Computer Problem 10.2.4) Plot the interpolating function for $P_n(t)$ on $[0,1]$ in Computer Problem 10.2.3, along with the data points and $f(t) = e^t$ for

## (a) $n = 16$

```python
# import needed libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Magic function to make matplotlib inline; other style specs must come AFTER
%matplotlib inline

# This enables high resolution PNGs.
%config InlineBackend.figure_formats = {'png', 'retina'}

# Some Seaborn settings for notebooks
rc = {'lines.linewidth': 2, 
      'axes.labelsize': 18, 
      'axes.titlesize': 18, 
      'axes.facecolor': 'DFDFE5'}
sns.set_context('notebook', rc=rc)
sns.set_style('darkgrid', rc=rc)
```

```python
n = 16

# create data (t, f(t)) data for f(t) = exp(t), 0 <= t <= 1
tdata = np.linspace(0,1,n,endpoint=False)
ydata = np.exp(tdata)

# compute the DFT of ydata
yh = np.fft.fft(ydata)/np.sqrt(n)
a = np.real(yh)
b = np.imag(yh)

# eqn 10.19, the trgonometric interpolating polynomial
# t is a vector
def pn(t,n,a,b,c,d):
    val = a[0] + a[n//2]*np.cos(n*np.pi*(t-c)/(d-c))
    for k in np.arange(1,n//2):
        val += 2*(a[k]*np.cos(2*np.pi*k*(t-c)/(d-c)) - b[k]*np.sin(2*np.pi*k*(t-c)/(d-c)))
    return val/np.sqrt(n)

# pp is the number of plot points
pp = 200

# tp are the independent variable values, equally spaced
tp = np.linspace(-0.05,1.0,pp,endpoint=True)

# c <= t <= d
c = 0
d = 1

# evaluate the poynomial
yp = pn(tp,n,a,b,c,d)

# plot the interpolating trigonomatric polynomil
plt.plot(tp, yp, 'b-', label='pn(t)')
# plot the data nodes
plt.plot(tdata, ydata, 'ro', label='data')
# plot exp(t)
plt.plot(tp, np.exp(tp), 'k-', label='exp(t)')
plt.xlabel('t')
plt.title('Trigonometric Interpolation (n = 16)')
plt.legend(loc='upper left')
plt.show()
```
